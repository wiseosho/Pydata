```python
from urllib.request import urlopen, build_opener
import time
import urllib.robotparser as Rp
import traceback, re, sys, os
import sqlite3
from bs4 import BeautifulSoup #beautifulsoup imports as bs4
import urllib.request as request
#함수화
contents = urlopen('http://blog.daum.net/').read()
crawler_name = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; de; rv:1.9.1.5) Gecko/20091102 Firefox/3.5.5'#'python_daum_crawler'
mainpage = 'http://blog.daum.net/'
mainpath = './data/'
```

```python
rp = Rp.RobotFileParser(mainpage + 'robots.txt')
rp.read()
```

```python
#rp.can_fetch('mycrawler', 'http://blog.daum.net')

def canFetch( url):
    #수집가능여부, robot.txt
    return rp.can_fetch(crawler_name, url)

def getContent(url, delay=1):
    #"웹문서 다운로드"
    time.sleep(delay)
    
    if not canFetch(url):
        print ('THis url can not be fetched by our crawler :', url)
        return None
    
    try:
        opener = build_opener()
        opener.addheaders = [('User-agent', crawler_name)]
        contents = opener.open(url).read()
    except:
        traceback.print_exec()
        return None
    return contents


def getArticleInfo(soup):
    #'http://blog.daum.net/'
    # URL href에 wrapping 되어있는 온전한 URL 리스트들을 리턴합니다.
    "Get article info inside the Daum Blog"
    rBlog = re.compile('.+blog.daum.net/\w+/\d+.*?')
    URLs = soup('a', {'href':rBlog})
    
    return [u.get('href').split('?')[0] for u in URLs ]

def getOwnArticles(contents):
    "해당 블로그에 보함되는 글의 목록을 가져옵니다."
    ret = []
    soup = BeautifulSoup(contents)
    rBlog = re.compile('.+BlogTypeView.+')
    for u in soup('a', {'href' : rBlog}):
        href = u.get('href')
        #Article주소에서 기사번호만 추출.
        #/_blog/BlogTypeView.do?blogid=05ASZ&articleno=15652780&_bloghome_menu=recenttext
        article = href.split('articleno=')[1].split('&')[0]
        #추출된 기사가 리스트에 없으면(?)
        if ret.count(article) < 1:
            ret.append(article)
            
    return ret


def getRedirectedURL(url):
    #From gatherNeighborInfo
    #-> blog.daum.net/UserID/
    #"본문에 해당하는 프레임의 URL을 읽어옴."
    contents = getContent(url)
    if not contents : return None
    
    #Redirect
    try:
        soup = BeautifulSoup(contents)
        frame = soup('frame')
        src = frame[0].get('src')
    except:
        src = None
        
    return src

def getBody(soup, parent):
    "Get body text from the relative url in the given soup contents"
    #Look for the iframe including body urls
    #Look for 'ArticleContentsView' from the given soup obj
    rSrc = re.compile('.+/ArticleContentsView.+')
    iframe = soup('iframe', {'src':rSrc})
    #If article contentsview exist, prefix with blog.daum.net and try to fetch body content again.
    #(Soup it again with real iframe URLs)
    #ArticleContentsView에 해당하는 첫번째 주소
    if len(iframe)>0:
        src = iframe[0].get('src')
        iframe_src = 'http://blog.daum.net'+src
        
        #add_headers referer -> 'parseArticle()의 newURL의 주소에서 지정' (<- parent 변수에서 상속)
        req = request.Request(iframe_src) #urllib.Request(url) for python 2.7
        req.addheaders('Referer', parent )
        body = urlopen(req).read()
        soup = BeautifulSoup(body)
        return str(soup.body)
    else:
        print ('NULL content')
        return ''
    
```

```python
def gatherNeighborInfo(soup):
    
    #이웃, 다녀간 블로거 리스트로부터 각각의 웹페잊에서의 getOwnArticle실행
    
    #Search for Daum Blogger address
    
    rBlog = re.compile('http://blog.daum.net/\w+')
    Neighbors = soup('a', {'href':rBlog})
    cnt = 0
    
    for n in Neighbors:
        url = n.get('href')
        blogname = url.split('/')[-1]
        
        if url and url.startswith('http://') and db.isCrawledURL(url) <1:
            #읽은 것으로 체크
            db.insertURL(url, 1)
            
            url2 = getRedirectedURL( url )
            #읽혀지지 않은 경우
            if not url2 : continue
            re_url = 'http://blog.daum.net' + url2
            body = getContent(re_url, 0)
            if body:
                for u in getOwnArticles(body):
                    #내글들의 목록을 Parse / blogtypeview 
                    fullpath = 'http://blog.daum.net/' + blogname+ '/' + u
                    
                    cnt += db.insertURL(fullpath)
    if cnt >0 : print ('%d neighbor articles inserted' % cnt)              
```

```python
def parseArticle(url):
    #"Parse the URL and Store"
    
    #Get blodID and ArticleID
    article_id = url.split('/')[-1]
    blog_id = url.split('/')[-2]
    
    #Parse redirected URL ( url)
    newURL = getRedirectedURL(url)
    
    if newURL:
        try:
            #make blog directory
            os.mkdir(mainpath + blogid)
            
        except:
            #ignore whil making directory
            pass
        
        newURL = 'http://blog.daum.net'+newURL
        contents = getContent(newURL, 0)
        if not contents:
            print ('NULL Contents...')
            
            db.updateURL(url, -1)
            return
        
        #Parse HTML
        soup = BeautifulSoup(contents)
        
        #Redirected URL 안에서 Parsing할 웹페이지가 가지는 다른 Article Information DB에 insert.
        # 내 블로그내 정보('BlogTypeView'), 이웃블로그. (blog.daum.net/\+w?)
        gatherNeighborInfo(soup)
        
        #db에 삽입.

        n = 0
        for u in getArticleInfo(soup):
            n += db.insertURL( u )
        if n>0 : print ('inserted %d urls from %s' %(n, url))
        
        #get title from the soup(body)
        sp = contents.find('<title>')
        if sp > -1:
            ep = contents[sp+7:].find('</title>')
            title = contents[sp+7:sp+ep+7].decode()
        else:
            title = ''
            
        #run get body
        #search for ArticleContentsView string
        contents = getBody(soup, newURL)
        
        #remove script
        
        pStyle = re.compile('<style(.*?)>(.*?)</style(.*?)', re.IGNORECASE | re.MULTILINE | re.DOTALL)
        contents = pStyle.sub('',contents)
        pStyle = re.compile('<script(.*?)>(.*?)<script(.*?)', re.IGNORECASE | re.MULTILINE | re.DOTALL)
        contents = pStyle.sub('',contents)
        pStyle = re.compile('<(.*?)>', re.IGNORECASE | re.MULTILINE | re.DOTALL)
        contents = pStyle.sub("", contents)
        
        #Store into text file
        
        ftxt = open(mainpath+blog_id + '/' + article_id + '.txt',  'w')
        ftxt.write(title + '\n')
        ftxt.write(contents)
        ftxt.close()
        
        #UPdate db for check
        db.updateURL(url)
        
        
        
    else: # newURL 이 존재하지 않는 경우 soup('frame')[0]('src')의 내용이 없는 경우. iframe 이 없는 경우
        print ('err in parseArticle, newURL does not exist, Invalid blog Article...')
        #db에 업데이트
        db.updateURL(url, -1)
        
    
```
