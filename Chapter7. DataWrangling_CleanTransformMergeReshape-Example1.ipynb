```python
nutrients = []

for rec in db:
    fnuts = DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)

nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients
```

```python
#Boolean Sum of Duplicates
#Any of them duplicated?

nutrients.duplicated().sum()
```

```python
nutrients[nutrients.duplicated()]
```

```python
#Drop false for duplicated nutrients
nutrients_uniq = nutrients.drop_duplicates()
```

```python
info
```

```python
col_mapping = {'description' : 'food',
              'group' : 'fgroup'}

info = info.rename(columns= col_mapping, copy = False)
info.head()
```

```python
col_mapping = {'description' : 'nutrient',
               'group' : 'nutgroup'}
nutrients_uniq.rename(columns=col_mapping, inplace=True)

nutrients_uniq.head()
```

```python
ndata = pd.merge(nutrients_uniq, info, on='id',how = 'outer')

ndata.head()
```

```python
#Extract axis with specified 'idx' value, default : axis 0, row / axis =1, columns

get_maximum = lambda x: x.xs(x.value.idxmax())
get_minimum = lambda x: x.xs(x.value.idxmin())

#Get the rows(include all columns) with .value is the maximum(idxmax()) and reconstruc nto DataFrame

max_foods = by_nutrient.apply(get_maximum)[['value', 'food']]
```

```python
#foods with maximum nutrient values

max_foods.food = max_foods.food.str[:50]
```
