```python
from __future__ import division
from numpy.random import randn
import numpy as np
import os
import matplotlib.pyplot as plt
np.random.seed(12345)
plt.rc('figure', figsize=(10, 6))
from pandas import Series, DataFrame
import pandas
import pandas as pd
np.set_printoptions(precision=4, threshold=500)
pd.options.display.max_rows = 100
```

```python
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df2 = DataFrame({'key': ['a', 'b', 'd'],
                 'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
#inner join, missing data from df2 will be omitted
# key is the column with same name ('key')
pd.merge(df1, df2) # how = inner, outer, left, right
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'],
                 'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
#If the names of keys are different,
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1,df2, how = 'outer')
```

```python
#Many to Many merge
df1 = DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'b'],
                'data1' : range(6)})
df2 = DataFrame({'key' : ['a', 'b', 'a', 'b', 'd'],
                 'data2': range(5)})

df1
```

```python
df2
```

```python
#left direction
pd.merge(df1, df2, on ='key', how ='left')
#df2 does not have a data for key 'c' (row 6)
```

```python
#Right direction merge (refer to left one)
pd.merge(df1, df2, on = 'key', how = 'right')
#df1 (left) does not have a key value for 'd' (row10)
```

```python
#both of dataframe should have values
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
#overlapping columns, key2 overlap, _x, _y
pd.merge(left, right, on = 'key1')
```

```python
pd.merge(left, right, on = 'key1', suffixes=('_left','_right'))
```

```python
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                  'value': range(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how = 'outer')
```

```python
#left_on, right_on : by column
#left_index, right_index : by index

pd.merge(right1, left1, left_index= True, right_on='key')
```

```python
pd.merge(right1, left1, left_index= True, right_on='key', how = 'outer')
```
