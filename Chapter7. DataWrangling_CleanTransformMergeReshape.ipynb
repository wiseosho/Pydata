```python
#Many to Many merge
df1 = DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'b'],
                'data1' : range(6)})
df2 = DataFrame({'key' : ['a', 'b', 'a', 'b', 'd'],
                 'data2': range(5)})

df1
```

```python
df2
```

```python
#left direction
pd.merge(df1, df2, on ='key', how ='left')
#df2 does not have a data for key 'c' (row 6)
```

```python
#Right direction merge (refer to left one)
pd.merge(df1, df2, on = 'key', how = 'right')
#df1 (left) does not have a key value for 'd' (row10)
```

```python
#both of dataframe should have values
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'outer')
```
