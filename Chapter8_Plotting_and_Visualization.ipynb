```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
# parse_dates ->
data = pd.read_csv('./pydata-book/ch08/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
```

```python
spx['2007-01-01':].plot(ax=ax,style ='g--')
```

```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
```

```python
#annotate
#xy = target coordinate with arros directed
#xytext = test location
#Series.asof(idx) = value of idx
#spx.asof(datetime(2007, 10, 11))

anot_list = []
for date, label in crisis_data:
    anot_list.append(ax.annotate(label, xy = (date, spx.asof(date) + 0),
               xytext = (date, spx.asof(date)+500),
               arrowprops=dict(facecolor='black'),
               horizontalalignment = 'left', verticalalignment = 'top') )
```

```python
[x.remove() for x in anot_list]
```

```python
anot_list = []
for date, label in crisis_data:
    anot_list.append(ax.annotate(label, xy = (date, spx[date] + 50),
                                xytext = (date, spx[date] + 250),
                                arrowprops=dict(facecolor='black'),
                                horizontalalignment = 'right', verticalalignment = 'bottom'))
```

```python
xlim = ax.set_xlim(['1/1/2007','1/1/2011'])
ylim = ax.set_ylim(auto=True)
title = ax.set_title('Important dates in 2008-2009 financial crisis')
```
