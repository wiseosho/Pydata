```python
#4. 사이클시간. Cycle time: 동일발 접촉시간차 (stride length: 활보장, Cycle time 동안 이동거리)
CycleTime = Gait.time.left.toeIn.diff(1).shift(-1)
CycleTimeR= Gait.time.right.toeOut.diff(1).shift(-1)
#11. 활보장. stride length: 동일 발뒤꿈치 이동간 Vector distance.
StrideLength = Gait.space.left.toeIn.diff(1).shift(-1)
StrideLengthR = Gait.space.right.toeOut.diff(1).shift(-1)
#12. 활보속도. stride velocity : Stride length / Cycle time
StrideVelocity = StrideLength / CycleTime
StrideVelocityR = StrideLengthR / CycleTimeR

print (StrideVelocity, StrideVelocityR)
```
